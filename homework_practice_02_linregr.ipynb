**Вопрос (0.1 балл)**: **(?)**Что можно сказать о целевой переменной по гистограмме её значений?

**Ответ**:
Первое, что бросается в глаза, это собранность большинства объектов по логарифму таргета в точке моды, а именнт, распредление таргета имеет большое значение эксцесса. Однако вместе с этим по первой (грубой) гистограмме можно увидеть, что у этого распределния тяжелые хвосты. А потому в выборке будут часто встречаться объекты с большими логарифмами времени, хотя в то же время модель будет очечаться в большинстве своя на средних показателях в районе $10^2, 10^3$

В соревновании в качестве метрики качества использовалось RMSLE:
$$\text{RMSLE}(X, y, a) = \sqrt{\frac{1}{\ell}\sum_{i=1}^{\ell} \big(\log{(y_i + 1)} - \log{(a(x_i) + 1)}\big)^2}$$

**Вопрос (0.1 балл)**: **(?)**Как вы думаете, почему авторы соревнования выбрали именно RMSLE, а не RMSE?

**Ответ:** Рассмторим для начала $\text{MSE}$ -- такую функцию эмпирического риска мы хотим использовать, грубо говоря, чтобы сильнее наказывать при больших отклонениях предсказания от реального значения таргета.

  > Заметим, последующее возведение в квадратный корень не меняет задачу минимизации из условия выпуклости и монотонности двух этих функций

Однако в данной задаче такое поведение будет идти только во вред. Этот вывод можно сделать исходя из гистограммы. Легче увидеть на примере левой, "грубой гистграммы", что у распределения таргета тяжелые хвосты, а значит, что при относительно маленькой ошибке, квадрат дельта ошибки получается огромным на больших таргетах, если считать, что модель строит гиперплоскость (действительно -- модель, которую мы будем обучать, является линейной, и она приближает все значения единой гиперплоскостью. И хотя на малых значениях она может делать это крайне хорошо, с увеличением исследуемого реального таргета ошибка должна линейно накапливаться, и вот зорошая модель становится плохой из-за таких объектов). Исправить это дело должны логарифмы в функции $\text{RMSLE}$. При таком риске все еще остается возрастающая с квадратичной скоростью ошибка, но при этом относительная ошибка не так сильно штрафуется при больших значениях таргета и предсказания

На семинаре мы рассматривали несколько моделей линейной регрессии в `sklearn`, но каждая из них оптимизировала среднеквадратичную ошибку (MSE), а не RMSLE. Давайте проделаем следующий трюк: будем предсказывать не целевую переменную, а ее *логарифм*. Обозначим $\hat{y}_i = \log{(y_i + 1)}$ — модифицированный таргет, а $\hat{a}(x_i)$ — предсказание модели, которая обучалась на $\hat{y}_i$, то есть логарифм таргета. Чтобы предсказать исходное значение, мы можем просто взять экспоненту от нашего предсказания: $a(x_i) = \exp(\hat{a}(x_i)) - 1$.

**Доказательство**: Посмотрим на саму функцию эмпирического риска RMSLE:
$Q(X^l, y, a)=\sqrt{\frac{1}{\ell}\sum_{i=1}^{\ell} \big(\log{(y_i + 1)} - \log{(a(x_i) + 1)}\big)^2}$

Обозначим для нее задачу оптимизации: $Q(X^\ell, y, a)\longrightarrow min$
Однако заметим, что из условия монотонности функций (однозначности отображения) и монотонности $\sqrt{\bullet}$ мы можем переписать задачу оптимизации в виде

$Q^{\prime}(X^l, y, a)=\frac{1}{\ell}\sum_{i=1}^{\ell} \big(\log{(y_i + 1)} - \log{(a(x_i) + 1)}\big)^2$

Точно так же заметим, что $\log(\bullet)$ так же является монотонной и выпуклой функцией, что похволяет нам построить взаимообратное соответсвие между задачами по минизации $Q^{\prime}$ и $\hat{Q}$, где $\hat{Q}$ получено из $Q^{\prime}$ заменой $\hat{y}_i = \log{(y_i + 1)}$, $\hat{a}(x_i) = \log{(a(x_i) + 1)}$.

Тогда, очевидно, мы переходим к задаче оптимизации задачи MSE
$$\hat{Q}=\frac{1}{\ell}\sum_{i=1}^{\ell} \big(\hat{y}_i - \hat{a}(x_i)\big)^2\longrightarrow min$$

И из минимизации первого функционала будет идти и минизация второго, а дальше и третьего $\hat{Q}$, и наборот.

**(0.1 балл)** Чтобы иметь некоторую точку отсчета, давайте посчитаем значение метрики MSE при наилучшем константном предсказании:

```python
np.random.seed(666) # адский сатана
```

```python
from sklearn.model_selection import train_test_split

dataset_target = data['log_trip_duration']
dataset = data.drop(columns=['log_trip_duration'])

x_train, x_test, y_train, y_test = train_test_split(dataset, dataset_target, test_size=0.2)
```

```python
from sklearn.metrics import mean_squared_error
from sklearn.linear_model import LinearRegression

# сделаем самую простую линейную регрессию

np.ones((x_train.shape[0])).reshape((-1, 1)).shape, y_train.shape
model = LinearRegression(fit_intercept=False)
model.fit(np.ones((x_train.shape[0])).reshape((-1, 1)), y_train)
y_train_pred = model.predict(np.ones((x_train.shape[0])).reshape((-1, 1)))
y_test_pred = model.predict(np.ones((x_test.shape[0])).reshape((-1, 1)))

linear_regr_error_train = mean_squared_error(y_train, y_train_pred)
linear_regr_error_test = mean_squared_error(y_test, y_test_pred)

# теперь же приблизим средним арифметическим

target_mean = y_train.mean()
dummy_error_train = mean_squared_error(y_train, np.array([target_mean] * y_train.shape[0]))
dummy_error_test = mean_squared_error(y_test, np.array([target_mean] * y_test.shape[0]))

# сравним результаты
print( '==========================  обучение ==========================')
print(f'Ошибка линейной регрессии на обучении:       {linear_regr_error_train}')
print(f'Ошибка среднего арифметического на обучении: {dummy_error_train}')
print( '============================  тест ============================')
print(f'Ошибка линейной регрессии на тесте:          {linear_regr_error_test}')
print(f'Ошибка среднего арифметического на тесте:    {dummy_error_test}')
print()
print(f'Коэффициент в линейной регрессии: {model.coef_[0]}, среднее арифметическое: {target_mean}')
```

Получили приближение примерно в 0.63 как на обучении, так и на тесте.
Осталось заметить, что линейная регрессия действительно сходится к ниболее точному решению для константного случая -> среднему от таргета

```python
def show_circles_on_map(data, latitude_column, longitude_column, color, *, point_list=[]):
    """
    The function draws map with circles on it.
    The center of the map is the mean of coordinates passed in data.

    data: DataFrame that contains columns latitude_column and longitude_column
    latitude_column: string, the name of column for latitude coordinates
    longitude_column: string, the name of column for longitude coordinates
    color: string, the color of circles to be drawn
    """

    location = (data[latitude_column].mean(), data[longitude_column].mean())
    m = folium.Map(location=location)

    if point_list:
      for point in point_list:
        folium.Circle(
              radius=10,
              location=(point[0], point[1]),
              color=color,
              fill_color=color,
              fill=True
          ).add_to(m)
    else:
      for _, row in data.iterrows():
          folium.Circle(
              radius=10,
              location=(row[latitude_column], row[longitude_column]),
              color=color,
              fill_color=color,
              fill=True
          ).add_to(m)

    return m
```

**(0.1 балла)** Как мы все прекрасно помним, $t = s / v_{\text{ср}}$, поэтому очевидно, что самым сильным признаком будет расстояние, которое необходимо проехать. Мы не можем посчитать точное расстояние, которое необходимо преодолеть такси, но мы можем его оценить, посчитав кратчайшее расстояние между точками начала и конца поездки. Чтобы корректно посчитать расстояние между двумя точками на Земле, можно использовать функцию `haversine`. Посчитайте кратчайшее расстояние для объектов и запишите его в колонку `haversine`:

```python
!pip install haversine
from haversine import haversine
```

```python
data2 = data1

def get_haversine(row):
  return haversine((row['pickup_latitude'], row['pickup_longitude']), (row['dropoff_latitude'], row['dropoff_longitude']))

data2['haversine'] = data2.apply(lambda x: get_haversine(x), axis=1)
data2
```

**(0.1 балла)** Так как мы предсказываем логарифм времени поездки и хотим, чтобы наши признаки были линейно зависимы с этой целевой переменной, нам нужно логарифмировать расстояние: $\log t = \log s - \log{v_{\text{ср}}}$. Запишите логарифм `haversine` в отдельную колонку:

```python
data2['log_haversine'] = np.log1p(data2.haversine)
```

**(0.2 балла)** Убедимся, что логарифм расстояния лучше коррелирует с нашим таргетом, чем просто расстояние:

```python
corr_target_haversine = data2.haversine.corr(data2.log_trip_duration)
corr_target_log_haversine = data2.log_haversine.corr(data2.log_trip_duration)

print(f'Коэффициент корреляции Пирсона между таргетом и расстоянием:           {corr_target_haversine}')
print(f'Коэффициент корреляции Пирсона между таргетом и логарифмом расстояния: {corr_target_log_haversine}')
```

Как легко увидеть, коэффициент корреляции между логарифмом расстояния и таргетом больше, чем между таргетом и просто расстоянием. Следовательно log_haversine, как новый признак лучше описывает поведение таргета

**Задание 9 (1 балл)**. Сейчас мы почти что не используем сами значения координат. На это есть несколько причин: по отдельности рассматривать широту и долготу не имеет особого смысла, стоит рассматривать их вместе. Во-вторых, понятно, что зависимость между нашим таргетом и координатами не линейная. Чтобы как-то использовать координаты, можно прибегнуть к следующему трюку: обрамим область с наибольшим количеством поездок прямоугольником (как на рисунке). Разобьем этот прямоугольник на ячейки. Каждой точке сопоставим номер ее ячейки, а тем точкам, что не попали ни в одну из ячеек, сопоставим значение -1.

Напишите трансформер, который сначала разбивает показанную на рисунке область на ячейки, а затем создает два признака: номер ячейки, в которой началась поездка, и номер ячейки, в которой закончилась поездка. Количество строк и столбцов выберите самостоятельно.

Обратите внимание, что все вычисления должны быть векторизованными, трансформер не должен модифицировать передаваемую ему выборку inplace, а все необходимые статистики (если они вдруг нужны) нужно считать только по обучающей выборке в методе `fit`:

```python
from sklearn.base import BaseEstimator, TransformerMixin

class MapGridTransformer(BaseEstimator, TransformerMixin):
  __slots__ = ('quantile_method', 'long_start', 'long_end', 'lat_start', 'lat_end', 'net_cells_x', 'net_cells_y')

  def __init__(self, *, quantile_method=False, net_cells_x=3, net_cells_y=3):
    self.net_cells_x = net_cells_x
    self.net_cells_y = net_cells_y
    self.quantile_method = quantile_method
    if not quantile_method:
      self.lat_start, self.long_start = (40.7096937850217, -73.94142544670183)
      self.lat_end, self.long_end = (40.817172620570354, -74.01207914929617)

  def fit(self, X, y=None):
    if self.quantile_method:
      X_without_abnormal_zones = X.loc[(X['start_JFKennedy'] == 0) & (X['end_JFKennedy'] == 0) & (X['start_LGuardia'] == 0) & (X['end_LGuardia'] == 0)]
      self.lat_start = pd.concat([X_without_abnormal_zones.pickup_latitude, X_without_abnormal_zones.dropoff_latitude], axis=0).quantile(0.05)
      self.lat_end = pd.concat([X_without_abnormal_zones.pickup_latitude, X_without_abnormal_zones.dropoff_latitude], axis=0).quantile(0.95)
      self.long_end = pd.concat([X_without_abnormal_zones.pickup_longitude, X_without_abnormal_zones.dropoff_longitude], axis=0).quantile(0.05)
      self.long_start = pd.concat([X_without_abnormal_zones.pickup_longitude, X_without_abnormal_zones.dropoff_longitude], axis=0).quantile(0.95)

    return self

  def transform(self, X):
    sectorized_X = X.copy()
    lat_delta = (self.lat_end - self.lat_start) / self.net_cells_x
    long_delta = (self.long_start - self.long_end) / self.net_cells_y

    def _get_sector(lat, long):
      if not (self.lat_start < lat < self.lat_end) or \
         not (self.long_end < long < self.long_start):
         return -1

      _x = int((lat - self.lat_start) / lat_delta)
      _y = int((long - self.long_end) / long_delta)
      return _y * self.net_cells_x + _x

    sectorized_X['sector_pickup'] = sectorized_X.apply(lambda row: _get_sector(row['pickup_latitude'], row['pickup_longitude']), axis=1)
    sectorized_X['sector_dropoff'] = sectorized_X.apply(lambda row: _get_sector(row['dropoff_latitude'], row['dropoff_longitude']), axis=1)

    return sectorized_X
```

```python
from sklearn.linear_model import Ridge
from sklearn.preprocessing import OneHotEncoder, StandardScaler
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline

dataset_target = data4['log_trip_duration']
dataset = data4.drop(columns=['log_trip_duration', 'store_and_fwd_flag', 'passenger_count', 'vendor_id'])

x_train, x_test, y_train, y_test = train_test_split(dataset, dataset_target, test_size=0.3)

categorical = ['weekday', 'month', 'hour']
numeric = ['year_day', 'log_haversine']
coordinates = ['pickup_longitude', 'pickup_latitude', 'dropoff_longitude', 'dropoff_latitude', 'start_JFKennedy', 'end_JFKennedy', 'start_LGuardia', 'end_LGuardia']

transformer = ColumnTransformer([
    ('ohe', OneHotEncoder(handle_unknown="ignore"), categorical),
    ('scaling', StandardScaler(), numeric),
    ('map_nesh', MapGridTransformer(net_cells_x=3, net_cells_y=4), coordinates)
])

pipeline = Pipeline(steps=[
    ('ohe_and_scaling', transformer),
    ('regression', Ridge())
])
model = pipeline.fit(x_train, y_train)
y_pred_train = model.predict(x_train)
y_pred_test = model.predict(x_test)
```

Давайте попробуем сделать следующее: посчитаем, сколько раз встречается каждая возможная пара этих признаков в нашем датасете и выберем 100 самых частых пар. Закодируем поездки с этими частыми парами как категориальный признак, остальным объектам припишем -1. Получается, что мы закодировали, откуда и куда должно было ехать такси.

Также можете придумать ещё какой-нибудь способ сделать признаки про маршрут. Если эти признаки будут давать хороший прирост в качестве, то за это могут быть даны дополнительные бонусные баллы **(до 1 бонусного балла)**

```python
class MapGridTransformerModified(BaseEstimator, TransformerMixin):
  __slots__ = ('quantile_method', 'long_start', 'long_end', 'lat_start', 'lat_end', 'net_cells_x', 'net_cells_y')

  def __init__(self, *, quantile_method=False, net_cells_x=3, net_cells_y=3):
    self.net_cells_x = net_cells_x
    self.net_cells_y = net_cells_y
    self.quantile_method = quantile_method
    if not quantile_method:
      self.lat_start, self.long_start = (40.7096937850217, -73.94142544670183)
      self.lat_end, self.long_end = (40.817172620570354, -74.01207914929617)

  def fit(self, X, y=None):
    if self.quantile_method:
      X_without_abnormal_zones = X.loc[(X['start_JFKennedy'] == 0) & (X['end_JFKennedy'] == 0) & (X['start_LGuardia'] == 0) & (X['end_LGuardia'] == 0)]
      self.lat_start = pd.concat([X_without_abnormal_zones.pickup_latitude, X_without_abnormal_zones.dropoff_latitude], axis=0).quantile(0.05)
      self.lat_end = pd.concat([X_without_abnormal_zones.pickup_latitude, X_without_abnormal_zones.dropoff_latitude], axis=0).quantile(0.95)
      self.long_end = pd.concat([X_without_abnormal_zones.pickup_longitude, X_without_abnormal_zones.dropoff_longitude], axis=0).quantile(0.05)
      self.long_start = pd.concat([X_without_abnormal_zones.pickup_longitude, X_without_abnormal_zones.dropoff_longitude], axis=0).quantile(0.95)

    return self

  def transform(self, X):
    sectorized_X = X.copy()
    lat_delta = (self.lat_end - self.lat_start) / self.net_cells_x
    long_delta = (self.long_start - self.long_end) / self.net_cells_y

    def _get_sector(lat, long):
      if not (self.lat_start < lat < self.lat_end) or \
         not (self.long_end < long < self.long_start):
         return -1

      _x = int((lat - self.lat_start) / lat_delta)
      _y = int((long - self.long_end) / long_delta)
      return _y * self.net_cells_x + _x

    sectorized_X['sector_pickup'] = sectorized_X.apply(lambda row: _get_sector(row['pickup_latitude'], row['pickup_longitude']), axis=1)
    sectorized_X['sector_dropoff'] = sectorized_X.apply(lambda row: _get_sector(row['dropoff_latitude'], row['dropoff_longitude']), axis=1)

    # добавляем новую фичу
    pairs = sectorized_X.groupby(by=['sector_pickup', 'sector_dropoff'], as_index=False).size().sort_values(by=['size'], ascending=False, ignore_index=True).values
    filtered_pairs = pairs[np.all(pairs[:, :2] >= 0, axis=1)][:100, [0, 1]] # stackoverflow driven development
    sectorized_X['way'] = -1
    for i, pair in enumerate(filtered_pairs):
      sectorized_X.loc[(sectorized_X.sector_pickup == pair[0]) & (sectorized_X.sector_dropoff == pair[1]), 'way'] = i

    return sectorized_X
```

```python
class ManhettenModifier(BaseEstimator, TransformerMixin):
  __slots__ = ('A_inv', 'Oxy', 'Oy', 'Ox')

  def __init__(self):
    self.Oxy = np.array([40.75208909164287, -73.99348346751934])
    self.Oy = np.array([40.80789367385267, -73.95286540299928])
    self.Ox = np.array([40.74384074156228, -73.9735837414767])

    f1 = (self.Oy - self.Oxy) / np.linalg.norm(self.Oy - self.Oxy)
    f2 = (self.Ox - self.Oxy) / np.linalg.norm(self.Ox - self.Oxy)
    self.A_inv = np.linalg.inv(np.hstack([f1.reshape((-1, 1)), f2.reshape((-1, 1))]))

  def fit(self, X, y=None):
    return self

  def transform(self, X):
    Y = X.copy()
    y_start = self.A_inv[0] @ np.array([(X['pickup_latitude'] - Oxy[0]), (X['pickup_longitude'] - Oxy[1])])
    x_start = self.A_inv[1] @ np.array([(X['pickup_latitude'] - Oxy[0]), (X['pickup_longitude'] - Oxy[1])])
    y_end = self.A_inv[0] @ np.array([(X['dropoff_latitude'] - Oxy[0]), (X['dropoff_longitude'] - Oxy[1])])
    x_end = self.A_inv[1] @ np.array([(X['dropoff_latitude'] - Oxy[0]), (X['dropoff_longitude'] - Oxy[1])])
    Y['manhetten_range'] = np.abs(y_start - y_end) + np.abs(x_start - x_end)
    Y['log_manhetten_range'] = np.log1p(data7['manhetten_range'])
    return Y
```
